```python
#Creating Price_label

def price_label(row):
    if row['prices']<= 5:
        return "Low Range Product"
    elif (row['prices']>5) and (row['prices']<=15):
        return "Mid Range Product"
    elif row['prices']>15:
        return "High Range Product"
    else: return np.nan
```

```python
df_subset['price_range'] = df_subset.apply(price_label, axis=1)
```

```python
#Creating the Busiest Day column

result = []

for value in df_merged['order_dow']:
    if value == 0:
        result.append("Busiest Day")
    elif value == 4:
            result.append("Least Busy")
    else:
            result.append("Regularly Busy")
    

```

```python
result
```

```python
df_merged['Busiest Day'] = result
df_merged['Busiest Day'].value_counts(dropna = False)
```

```python
df_merged.shape
```

```python
# Step 2 - Creating the two busiest and slowest days column

result = []
for value in df_merged['order_dow']:
    if value == 0 or value == 1:
        result.append("Busiest Day")
    elif value == 4 or value == 3:
        result.append("Slowest Day")
    else:
        result.append("Regular Day")
```

```python
result
```

```python
df_merged['Busiest & Slowest Days'] = result
df_merged['Busiest & Slowest Days'].value_counts(dropna = False)
```

```python
df_merged['order_hour_of_day'].value_counts(dropna=False)
```

```python
#Creating the column busiest period of the day

result=[]
for value in df_merged['order_hour_of_day']:
    if value >= 9 and value <= 17:
        result.append('Most Orders')
    elif value >= 0 and value <= 6:
        result.append('Fewest Orders')
    else:
        result.append('Average Orders')
    
```

```python
result
```

```python
df_merged['Busiest Period of the Day'] = result
df_merged['Busiest Period of the Day'].value_counts(dropna = False)
```
